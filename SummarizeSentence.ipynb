```python
import csv
from pythainlp.summarize import summarize_text
text = []

with open('Raw_Data.csv', newline='', encoding='utf-8') as File:  
    reader = csv.reader(File)
    for row in reader:
        text.append(row[0])
```

```python
from pythainlp.summarize import summarize_text

for row in text:
    print(summarize_text(text=row,n=3,engine='frequency'))
```
